# SKIN DISEASE IDENTIFICATION USING IMAGE ANALYSIS

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Convolution2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Flatten
```

```python
model = Sequential()
```

Adding Convolutional layers

```python
model.add(Convolution2D(32,(3,3),input_shape = (64,64,3), activation = 'relu'))
```

```python
model.add(Convolution2D(32,(3,3),input_shape = (64,64,3), activation = 'relu'))
```

Adding Maxpooling layer

```python
model.add(MaxPooling2D(pool_size = (2,2)))
```

Add Flattening layer

```python
model.add(Flatten())
```

```python
model.summary()
```

Adding hidden layers

```python
model.add(Dense(128, activation = 'relu', kernel_initializer= 'uniform'))
```

```python
model.add(Dense(128, activation = 'relu', kernel_initializer= 'uniform'))
```

```python
model.add(Dense(64, activation = 'relu', kernel_initializer= 'uniform'))
```

```python
model.add(Dense(64, activation = 'relu', kernel_initializer= 'uniform'))
```

Adding output layer

```python
model.add(Dense(5, kernel_initializer='uniform', activation='softmax')) 
```

Preprocessing the data

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255, shear_range = 0.2, zoom_range = 0.2, height_shift_range = 0.2, width_shift_range = 0.2, horizontal_flip = True, vertical_flip = True)
test_datagen = ImageDataGenerator(rescale = 1)
```

Loading the train and test data

```python
x_train = train_datagen.flow_from_directory(r'C:\Users\nisha\Documents\Skin-disease-identification-system\Skin_Diseases\train', target_size = (64, 64), batch_size = 32, class_mode = 'categorical')
x_test = test_datagen.flow_from_directory(r'C:\Users\nisha\Documents\Skin-disease-identification-system\Skin_Diseases\test', target_size = (64, 64), batch_size = 32, class_mode = 'categorical')
```

```python
x_train.class_indices
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

Fitting the data

```python
model.fit(x_train, steps_per_epoch = 100, epochs = 200, validation_data = x_test, validation_steps = 63)
```

Saving the model

```python
model.save("Skin_Diseases.h5")
```
